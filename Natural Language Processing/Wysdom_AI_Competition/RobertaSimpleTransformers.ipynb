```python
# prepare cross validation
n=5
kf = KFold(n_splits=n, random_state=42, shuffle=True)
results = []

for train_index, val_index in kf.split(df_mock):
    # splitting Dataframe (dataset not included)
    train_df = df_mock.iloc[train_index]
    val_df = df_mock.iloc[val_index]
    
    model = ClassificationModel('roberta', 'roberta-base', num_labels=37,use_cuda=True, args={'learning_rate':1e-5, 'num_train_epochs': 1, 'reprocess_input_data': True, 'overwrite_output_dir': True}) 
    model.train_model(train_df)
    result, model_outputs, wrong_predictions = model.eval_model(val_df, acc=sklearn.metrics.accuracy_score)
    print(result['acc'])
    results.append(result['acc'])
```

```python
from sklearn.metrics.cluster import adjusted_mutual_info_score
# validate the model 
result, model_outputs, wrong_predictions = model.eval_model(val_df, acc=sklearn.metrics.adjusted_rand_score)
print(result['acc'])
# append model score
results.append(result['acc'])
```
